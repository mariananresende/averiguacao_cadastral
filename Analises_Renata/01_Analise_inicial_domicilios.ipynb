# Funcoes

## Print message

```python
def print_message(msg):
    print(color.BOLD + color.BLUE + '\n' + msg + color.END)
    print(color.BOLD + color.BLUE + 50*'*' + color.END)
```

## Checar características de um dataframe

```python
def checar_df(df):
    print(color.BOLD  + color.RED + 'Informações do dataframe: ' + color.END)
    df.info(verbose=True, show_counts=True)
    
    print(color.BOLD + color.RED + "\nNúmero de observacoes na base de dados: " + color.END, df.shape[0])
    print('Total de ', df.shape[1], 'Colunas: ', df.columns.values.tolist())
    
    print(color.BOLD + color.RED + "\nNúmero de valores unicos (Não inclui NaN): " + color.END)
    display(df.nunique(dropna=True).to_frame(name='Contagem Unicos').sort_index())
    
    print(color.BOLD + color.RED + "Número de valores nulos:" + color.END)
    display(df.isnull().sum().sort_values(ascending=False).to_frame(name='Contagem Nulos').sort_index())
```

```python
# passar como parâmetro a lista de colunas para as quais se deseja obter os valores únicos existentes
def checar_val_colunas_df_lista(df, lista_colunas):
    print(color.BOLD + color.RED + "Contagem de valores unicos (inclui NaN) e apresentação da lista de valores:  \n"  
      + color.END)
    for column in lista_colunas:
        print(column, ': ', len(df[column].unique()),'valores únicos \n',
              df[column].sort_values().unique().tolist(), '\n')
```

```python
# passar como parâmetro o número máximo de valores únicos (n_unique)
# Ex. mostrar colunas e seus valores únicos apenas quando a coluna tem acima de N valores únicos
def checar_val_colunas_df_nunique(df, n_unique):
    df_count_unique = df.nunique(dropna=True).to_frame(name='Contagem Unicos').sort_values(by='Contagem Unicos')
    lista_colunas = df_count_unique[df_count_unique['Contagem Unicos'] >= n_unique].index
    
    print(color.BOLD + color.RED + "Contagem de valores unicos (inclui NaN) e apresentação da lista de valores: "  
      + color.END)
    print(color.BOLD + color.RED + '(acima de', n_unique, 'valores únicos)\n' + color.END)
    for column in lista_colunas:
        print(column, ': ', len(df[column].unique()),'valores únicos \n',
              df[column].sort_values().unique().tolist(), '\n')
```

## Contagens de valores de uma coluna

```python
def counts(df, col, n): 
    '''
    Calcula quantidade (total e percentual) de cada valor único da coluna.
    Informar os seguintes parâmetros:
      - df: dataframe
      - col: coluna de interesse 
      - n: os n top valores
    '''
    count_coluna = df[col].value_counts()
    perc_coluna = df[col].value_counts(normalize=True)*100
    print('Total de registros: ', df.shape[0])
    print('Total de registros de cada valor da coluna '+ col + ':', '(top', n,')')
    display(pd.DataFrame({'Total': count_coluna, 'Total em %': perc_coluna}).head(n))
```

## Criar campo Codigo e Nome da Faixa de Renda

```python
def define_cod_faixa_renda(x):
    '''
    PBZ: pobreza
    BRD: baixa renda
    ASM: acima de 1/2 salario mínimo
    '''
    meio_sm = vlr_salario_minimo_2018/2
    
    if x >=0 and x <= vlr_lim_pobreza: return 'PBZ'
    elif x > vlr_lim_pobreza and x <= meio_sm: return 'BRD'
    elif x > meio_sm: return 'ASM'
    else: return 'não informado'
```

```python
def define_nom_faixa_renda(x):
    '''
    PBZ: pobreza
    BRD: baixa renda
    ASM: acima de 1/2 salario mínimo
    '''
    meio_sm = vlr_salario_minimo_2018/2
    
    if x >=0 and x <= vlr_lim_pobreza: return 'Pobreza'
    elif x > vlr_lim_pobreza and x <= meio_sm: return 'Baixa Renda'
    elif x > meio_sm: return 'Acima de 1/2 salário mínimo'
    else: return 'não informado'
```

## Criar campos nomes a partir de dict

```python
def incluir_nomes(df, coluna, nova_coluna, mapa_nomes):
    print_message('Valores existentes')
    display(df[coluna].value_counts().reset_index())
    df[nova_coluna] = df[coluna].map(mapa_nomes)
    print_message('Após criação da nova coluna de nome')
    tmp = df.groupby([coluna, nova_coluna])['id_familia']\
            .count().reset_index(name='total')\
            .sort_values(by='total', ascending=False)
    display(tmp)
```

## Gerar boxplots - renda

```python
def boxplot_vlr_renda_hue(df, hue, palette):
    f, axes = plt.subplots(1, 1,  figsize=(20, 10))
    g = sns.boxplot(x='nome_regiao_ibge', 
                    y='vlr_renda_media_fam', 
                    hue=hue, 
                    data=df, palette=palette)

    media = df['vlr_renda_media_fam'].mean()
    g.axhline(media, linestyle='--', linewidth = 1.5, color='r', label='Renda Media')

    major_ticks = np.arange(0, 3000, 500)
    minor_ticks = np.arange(0, 3000, 100)
    axes.set_yticks(major_ticks)
    axes.set_yticks(minor_ticks, minor=True)
    axes.grid(which='both', axis='y')

    xlocs, xlabels = plt.xticks()
    ylocs, ylabels = plt.yticks()
    plt.setp(xlabels, fontsize=16, rotation=0)
    plt.setp(ylabels, fontsize=16)
    axes.set(xlabel=''); 
    axes.set_ylabel('Renda Familiar', fontsize=16) 

    axes.set_title('Renda Familiar por Região e %s' %hue, fontsize=20, pad=30)
    axes.legend(fontsize=16, loc='upper center', ncol=5)
    plt.show()
```

## Variaveis independentes com a Target

```python
def corr_spearman(df, lista_col, col_top, k):
    k = k # número de variáveis
    corrmat = abs(df[lista_col].corr(method='spearman')) # correlação de spearman
    cols = corrmat.nlargest(k, col_top).index # o k-ésimo maior valor
    cm = np.corrcoef(df[cols].values.T) # calcula a correlação
    
    sns.set(font_scale=1.0)
    f, ax = plt.subplots(figsize=(20, 12))
    
    mask = np.zeros_like(cm) 
    mask[np.triu_indices_from(mask)] = True 
    sns.set_style("white")
    hm = sns.heatmap(cm, cbar=True, 
                     annot=True, 
                     square=True, 
                     fmt='.2f',
                     annot_kws={'size': 10}, 
                     yticklabels=cols.values, 
                     xticklabels=cols.values, 
                     mask = mask)
    plt.show()
    return cm, cols, corrmat
```
