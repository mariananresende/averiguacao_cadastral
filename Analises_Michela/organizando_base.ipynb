```python
# FUNÇÃO PARA VERIFICAR A QUANTIDADE DE DADOS FALTANTES

def missing_data_info(data):
    missing_count = data.isnull().sum()
    missing_percent = 100 * data.isnull().mean()
    data_types = data.dtypes
    missing_info = pd.DataFrame({
        'Missing Count': missing_count,
        'Missing Percent': missing_percent,
        'Data Type': data_types
    })

    return missing_info.sort_values(by='Missing Percent', ascending=False)

# missing_data_info(df)
```

## Tratando características das pessoas

```python
# filtrando base
selected_columns = var_id + var_pes
df1 = df[selected_columns]
df1
```

```python
df1.columns
```

```python
# CRIANDO DF VARIÁVEIS PESSOAS ANALISADAS POR FAMILIA

# Calcular percentual pessoas com determinada característica por familia
df_var_caracterisca_pessoas = df1.groupby('id_familia').apply(
    lambda x: pd.Series({
        'qde_pessoas': x['qtde_pessoas'].iloc[0],  # mesma quantidade por id_familia
        'pct_mulheres': (x['cod_sexo_pessoa'] == 2).sum() / x['qtde_pessoas'].iloc[0],
        'pct_homens': (x['cod_sexo_pessoa'] == 1).sum() / x['qtde_pessoas'].iloc[0],
        'pct_criancas': (x['idade'] < 14).sum() / x['qtde_pessoas'].iloc[0],
        'pct_jovens_entre_15_29': ((x['idade'] > 14) & (x['idade'] <= 29)).sum() / x['qtde_pessoas'].iloc[0],
        'pct_idosos': (x['idade'] >= 60).sum() / x['qtde_pessoas'].iloc[0],
        'pct_deficientes': (x['cod_deficiencia_memb'] == 1).sum() / x['qtde_pessoas'].iloc[0]        
    }),
    include_groups=False
).reset_index()

df_var_caracterisca_pessoas
```

```python
# substituindo valores faltantes pela média, mediana ou outra função

def fill_missing_custom(df, columns, fill_func):
    """
    Substitui os dados faltantes de uma lista de colunas usando uma função personalizada.

    :param df: DataFrame
    :param columns: Lista de colunas nas quais os dados faltantes serão substituídos
    :param fill_func: Função personalizada para substituir os dados faltantes (ex: mean, median, etc.)
    :return: DataFrame com valores faltantes substituídos
    """
    for column in columns:
        fill_value = fill_func(df[column])  # Aplica a função personalizada (ex.: mediana)
        df[column] = df[column].fillna(fill_value)  # Substitui os valores nulos sem usar inplace

    return df

# SUBSTITUINDO PELA MÉDIA
# df_encoded = fill_missing_custom(df_encoded, ['qtd_comodos_domic_fam', 'qtd_comodos_dormitorio_fam'], lambda x: x.mean())

# SUBSTITUINDO PELA MEDIANA
df_encoded = fill_missing_custom(df_encoded, ['qtd_comodos_domic_fam', 'qtd_comodos_dormitorio_fam'], lambda x: x.median())
```

```python
# deletando linhas ausentes

def drop_rows_with_missing(data):
    """
    Remove todas as linhas que contêm valores faltantes no DataFrame.

    :param df: DataFrame
    :return: DataFrame sem linhas com valores faltantes.
    """
    # Remove as linhas que contêm valores nulos
    df_clean = data.dropna()

    return df_clean

# Aplicação
df_encoded = drop_rows_with_missing(df_encoded)

missing_data_info(df_encoded)
```
