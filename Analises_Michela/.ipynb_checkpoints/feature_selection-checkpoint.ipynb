# Script de seleção de features 




### Métodos Chi Quadrado

```python
# Importando as bibliotecas necessárias

import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold
```

```python
# FUNÇÃO PARA VERIFICAR A QUANTIDADE DE DADOS FALTANTES

def missing_data_info(data):
    missing_count = data.isnull().sum()
    missing_percent = 100 * data.isnull().mean()
    data_types = data.dtypes
    missing_info = pd.DataFrame({
        'Missing Count': missing_count,
        'Missing Percent': missing_percent,
        'Data Type': data_types
    })

    return missing_info.sort_values(by='Missing Percent', ascending=False)

# missing_data_info(df)
```

```python
# Carregando o dataset sem missing

pd.set_option('display.max_columns', 160)
pd.set_option('display.max_rows', 160)

df = pd.read_csv('bases/df_tratado.csv', sep=',')
print(f'Shape:', df.shape)

df
```

```python
# FEATURE SELECTION - MÉTODO CHI-QUADRADO

# Selecionando variáveis preditoras e target
df1 = df.drop('id_familia', axis=1)
X = df1.drop('classe_renda', axis=1)
y = df1['classe_renda']

# Selecionando duas variáveis com maior chi-quadrado
chi2_selector = SelectKBest(chi2, k=150) # MÉTODO CHI-QUADRADO

X_kbest = chi2_selector.fit_transform(X, y)

# Obtendo os scores e os nomes das variáveis
scores = chi2_selector.scores_
selected_features = X.columns[chi2_selector.get_support()]

# Criando um DataFrame com os nomes das variáveis e seus scores
features_scores_df = pd.DataFrame({
    'Feature': X.columns,
    'Score': scores
})

# Filtrando apenas as variáveis selecionadas
features_scores_df = features_scores_df[features_scores_df['Feature'].isin(selected_features)]

# Ordenando o DataFrame pelos scores em ordem decrescente
features_scores_df = features_scores_df.sort_values(by='Score', ascending=False)

features_scores_df
```

```python
# salvar o score das features
features_scores_df.to_excel('features_scores_df.xlsx', index=False)

```

```python
# FEATURES IRRELEVANTES

features_scores_df = features_scores_df.sort_values(by='Score', ascending=False)
irrelevant_features1 = features_scores_df[features_scores_df['Score'] < 1]['Feature'].tolist()
irrelevant_features5 = features_scores_df[features_scores_df['Score'] < 5]['Feature'].tolist()

print(f'Shape do DataFrame original: {df.shape}')

# Removendo colunas irrelevant_features1
dfa = df.drop(columns=irrelevant_features1)

print(f'Shape do DataFrame sem irrelevant_features1: {dfa.shape}')

# Removendo colunas irrelevant_features5
dfb = df.drop(columns=irrelevant_features5)

print(f'Shape do DataFrame sem irrelevant_features5: {dfb.shape}')
```

```python
# AJUSTANDO MODELO DE REGRESSÃO LOGISTICA COM TODAS AS FEATURES

# separando variáveis preditoras e variável alvo
df1 = df.drop('id_familia', axis=1)
X = df1.drop('classe_renda', axis=1)
y = df1['classe_renda']

# Separando dados em folds
skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)

# Criando o modelo
model = LogisticRegression(penalty='l2', solver='saga', max_iter=1000) 

# Avaliando o modelo
scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
print("Scores de precisão em cada fold:", scores)
print("Precisão média:", scores.mean())
```

```python
# AJUSTANDO MODELO DE REGRESSÃO LOGISTICA SEM FEATURES SCORE<1

# separando variáveis preditoras e variável alvo
X = dfa.drop('classe_renda', axis=1)
y = dfa['classe_renda']

# Separando dados em folds
skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)

# Criando o modelo
model = LogisticRegression(penalty='l2', solver='saga', max_iter=1500) 

# Avaliando o modelo
scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
print("Scores de precisão em cada fold:", scores)
print("Precisão média:", scores.mean())
```

```python
# AJUSTANDO MODELO DE REGRESSÃO LOGISTICA SEM FEATURES SCORE<5

# separando variáveis preditoras e variável alvo
X = dfb.drop('classe_renda', axis=1)
y = dfb['classe_renda']

# Separando dados em folds
skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)

# Criando o modelo
model = LogisticRegression(penalty='l2', solver='saga', max_iter=1500) 

# Avaliando o modelo
scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
print("Scores de precisão em cada fold:", scores)
print("Precisão média:", scores.mean())
```
