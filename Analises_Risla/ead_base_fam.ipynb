```python
mask_fam = np.triu(np.ones_like(df_fam.corr(), dtype=bool), k=1)
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(df_fam.corr(), annot=True, cmap='coolwarm', fmt='.2f', mask=mask_fam)
plt.title('Matriz de Correlação')
plt.show()
```

Temos duas correlações acima de 0.8 - portanto, são correlações fortes -, a saber: 
- qnt_comodos_dormitorio_fam X qnt_comodos_domic_fam = 0.81
- cod_banheiro_domic_fam X cod_agua_canalizada_fam = 0.83
